Append() --> Add new item in list. 
Insert() --> Add new item in list point location.
Pop()    --> Remove item in list at last position. 
Index()  --> Item calling index number. 
Join()   --> Conneting list items. 
Remove() --> Item deleting in list. 

```python
letter = ['a', 'b', 'c', 'd']
letter.append("e")
print(letter)
print("--"*42)
letter.insert(3,"ç")
print(letter)
print("--"*42)
letter.pop()                                   # Pop work like append but remove item.
print(letter)
print("--"*42)
print("Inex Number " ,letter.index("ç"))
print("--"*42)
paraf= "-".join(letter)
print(paraf,type(letter),"to",type(paraf))     # List changeing str variables. 
letter.remove(letter[3])
print("--"*42)
print(letter)
```

```python
_win=["Dec","Jan","Feb"]                    # Multi layer list sample.
_spr=["Mar","Apr","May"]
_sum=["Jun","Jul","Aug"]
_fal=["Sep","Oct","Nov"]
seasons=[_win,_spr,_sum,_fal]
print(seasons)
print("--"*42)
print(seasons[2])
print("--"*42)
print(seasons[2][1])
```

Dictionaries --> Making  Key : Value

```python
greetings = {
        "english": "hello",
        "japanese": "こんにちは",
        "german": "hallo",
        "hindi": "नमस्ते",
        "leet": "h3ll0" }

print(greetings)
print("--"*42)
print(greetings["german"])                  # Calling data.
print("--"*42)
print(list(greetings.items()))
print("--"*42)
print(list(greetings.keys()))
print("--"*42)
print(list(greetings.values()))
print("--"*42)
greetings["japanese"]= "Ohayou"             # Value update 
print(greetings)
print("--"*42)
greetings["turkish"]= "Selamın Aleyküm."    # Add new item
print(greetings)
print("--"*42)
del greetings["leet"]                       # Delete item.
print(greetings)
print("--"*42)
print("German in list : ","german" in greetings)  # Checking dictionary
print("--"*42)
print("leet in list : ","leet" in greetings)
print("--"*42)
```

Tuples --> Very similar like list. But use for fixed values. Tubles dont have append pop or like etc. Useing with ()
           Tuples just have 2 metod. Index() and Count()

```python
my_tuple=(1,4,3,4,5,6,7,4)

print(my_tuple)
print('Index:', my_tuple.index(4))    # Firsth ındex value
print('Count:', my_tuple.count(4))    # Count value quantty
```

Sets --> Adding multi data types but dont add similar values. Sets dont have index number. 

Sets Metods ; 

**remove()**: Remove set parameters

**add()**: Add new item in set

**difference()**: Two sets difference showing.

**discard()**: Remove item in set

**intersection()**: Two sets intersection show

**union()**: Showing two sets all parameter. 

```python
fruit = set(['apple','pear', 'orange','cherry', 'banana'])
print(fruit,"<--",type(fruit))
print("--"*42)
karakterler = "Veri Bilimi Bootcamp"
küme = set(karakterler)
print(küme)
```
