Functions

```python
def hip(kenar_1, kenar_2):
    
    hipotenus = ((kenar_1**2) + (kenar_2**2))**0.5 
    
    return hipotenus

print(hip(3,4))
print(hip(6,8))
print(hip(10,17.3))
```

```python
def kucuk_kareler_listeye(x):
    
    kucukler_listesi=[]
    for y in range(x):
        kucukler_listesi.append(y**2)
    
    return kucukler_listesi

print (kucuk_kareler_listeye(15))

print("--"*42)

def kucukler_listeye(x): # List comprehension
    
    kucukler_listeye=[y for y in range (x)]
    
    return kucukler_listeye

print (kucukler_listeye(10))
```

```python
def cift_listele(x):
    
    result = []
    for a in range(0,x):
        if a%2==0:
            result.append(a)
    return result

print(cift_listele(24))

print("--"*42)

# Lambda version

print(list(filter((lambda x: x%2==0), range(0,24))))

```

```python
# Map Sample

rakamlar = [1, 2, 3, 4, 5, 6, 7, 8, 9]

print(list(map((lambda x: x **2), rakamlar)))
```

```python
%%writefile example.py 
def make_list_in_range(x,y):
    
    new_list=[]
    for a in range(x+1,y):
        new_list.append(a)
    
    return new_list

def make_list_of_even(x,y):
    
    new_list=[]
    for a in range(x+1,y):
        if a%2==0:
            new_list.append(a)
    
    return new_list
```

```python
import example # Calling module
liste = example.make_list_in_range(1,20)
print(liste)
print("--"*42)
from example import make_list_in_range 
print(make_list_in_range(1,5))
print("--"*42)
import example as ex
print(ex.make_list_in_range(1,8))
```
