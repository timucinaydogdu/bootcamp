```python
with open ("davet.txt","r",encoding="utf8") as f:
    text_1 = f.read(29)  # read first 29 character
    text_2 = f.read(38)  # read continue start than 29 
    f.seek(0)            # use for cursor position
    text_3 = f.read(29)  
    position = f.tell()  # listening cursor potion
print(text_1)
print(text_2)
print(text_3)
print(position)
print("--"*42)
```

```python
with open ("davet.txt","r",encoding="utf8") as f:
    line=f.read()
print(line)
```

```python
with open ("davet.txt","r",encoding="utf8") as f:
    line=f.readline()
    rest_of_lines=f.readlines()
print(line)
print("--"*42)
print(rest_of_lines)
```
